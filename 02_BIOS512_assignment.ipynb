Use `geom_bar` to make a histogram of penguin counts by `year`.

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res=250)

p = ggplot(data=penguins, aes(x=year, fill=year))

p = p + geom_bar() + ylab("Number of Penguins")

p
```

Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width=6, repr.plot.height=6, repr.plot.res=250)

p = ggplot(data=penguins, aes(x=body_mass_g))

p = p + geom_bar(na.rm=TRUE)

p = p + facet_wrap(~year, ncol=1)

p = p + theme_light() + scale_x_continuous(label=comma) + xlab("Body Mass in Grams") + ylab("Number of Penguins")

p
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
options(repr.plot.width=6, repr.plot.height=6, repr.plot.res=250)

p = ggplot(data=penguins, aes(x=body_mass_g, y=sex))

p = p + geom_point(position = position_jitter(h = 0.2, w = 0), na.rm=TRUE, alpha=.5)

p = p + facet_wrap(~species, ncol=1) + theme_light()

p = p + scale_x_continuous(label=comma) + xlab("Body Mass in Grams")

p
```
