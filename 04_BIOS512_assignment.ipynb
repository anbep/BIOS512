```r
library("tidyverse")
```

```r
windturbines = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv',
        col_types=cols())

windturbines %>% head
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
df_wind = windturbines %>%
    mutate(province_territory = province_territory %>% fct_infreq() %>% fct_rev)
```

```r
df_wind %>% pull(province_territory) %>% levels
```

```r
p = ggplot(df_wind, aes(y=province_territory))

p = p + geom_bar()

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
df_wind2 = windturbines %>%
    mutate(province_territory = province_territory %>% fct_lump(n=5) %>% fct_infreq() %>%fct_relevel("Other", after=Inf) %>% fct_rev())
```

```r
df_wind2 %>% pull(province_territory) %>% levels
```

```r
p = ggplot(df_wind2, aes(y=province_territory))

p = p + geom_bar()

p
```
